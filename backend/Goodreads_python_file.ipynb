```python
path = '/Users/wangyunhuang/Documents/Projects/Goodreads/archive'
```

```python
files = [file for file in os.listdir(path) if file.endswith('csv')]
```

```python
books = pd.DataFrame()
```

```python
for file in files:
    current = pd.read_csv(os.path.join(path, file))
    books = pd.concat([books, current])
```

```python
books = books.drop(['PublishMonth', 'PublishDay'], axis = 1)
```

```python
high_rate = books[books['Rating'].astype(float) > 4.0]
high_rate
```

```python
# Show the number of languages of the books
high_rate['Language'].nunique()
```

```python
high_rate['Language'].unique()
```

```python
# I read in English, Chinese and French, list the books in these languages
high_rate_mylan = high_rate[high_rate['Language'].isin(['eng', 'en-US', 'fre', 'en-GB', 'en-CA', 'zho'])]
high_rate_mylan
```

```python
high_rate_mylan['RatingDistTotal'] = high_rate_mylan['RatingDistTotal'].str.extract('(\d+)')
```

```python
high_rate_mylan['RatingDistTotal'] = high_rate_mylan['RatingDistTotal'].astype(int)
```

```python
# Filter out books with counts of ratings less than 4000

high_rate_high_rater = high_rate_mylan[high_rate_mylan['RatingDistTotal'] > 4000]
high_rate_high_rater
```

```python
high_rate_high_rater = high_rate_high_rater.rename(columns = {'RatingDist5': 'NumOf5Star', 'RatingDist4': 'NumOf4Star',\
                                'RatingDist3': 'NumOf3Star','RatingDist2': 'NumOf2Star','RatingDist1': 'NumOf1Star'})
```

```python
columns_spliting_counts = ['NumOf5Star', 'NumOf4Star', 'NumOf3Star', 'NumOf2Star', 'NumOf1Star']
for col in columns_spliting_counts:
    high_rate_high_rater[col] = high_rate_high_rater[col].str.split(':').str[1] 
```
