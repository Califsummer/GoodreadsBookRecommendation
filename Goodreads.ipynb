```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
from langdetect import detect
```

```python
# Detect the language a book is in and tokenize it (for English and French)
def description_preprocessing(text):
    lang = detect(text)
    
    if lang in ['english', 'french']:
        stop_words = set(stopwords.words(lang))
    else:
        stop_words = set()
    
    words = word_tokenize(text)
    words = [word for word in words if word not in stop_words]
    
    text = ' '.join(words)
    
    return text

high_rate_high_rater['Description'] = high_rate_high_rater['Description'].apply(description_preprocessing)
```
